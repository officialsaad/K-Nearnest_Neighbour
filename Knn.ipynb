# KNN Classification 


## *Importing Libraries*

```python
import pandas as pd
```

### *Loading the dataset from excel files* 


```python
training_dataframe = pd.read_excel("training_data.xlsx")
training_dataframe.head()
```

```python
testing_dataframe = pd.read_excel("testing_data.xlsx")
testing_dataframe.head()
```

### *Data Cleaning*

```python
# Drop the rowno column
df = training_dataframe.drop('rowno', axis=1)

# Drop the rowno from testing dataframe

testing_dataframe = testing_dataframe.drop('rowno', axis=1)
```

```python
df.head(10)
```

```python
testing_dataframe.head()
```

### *Selection of the input and output variables*

```python
x = df[['f1','f2', 'f3','f4']]
y = df['class_labels']
```

```python
x.head()
```

```python
y.head()
```

### *Load and fit KNN Classifier*

```python
# Model and prediction 

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3, metric='minkowski')
knn.fit(x,y)

```

```python
t_x = testing_dataframe[['f1','f2','f3','f4']]
t_y = testing_dataframe[['class_label']]
```

```python
t_x.head()
```

```python
t_y.head()
```

```python
predictions =  knn.predict(t_x)
predictions
```

```python
print(predictions)
```

```python
# Calculate classification rate 
## two correct and one wrong prediction made by the model so calculate classification rate 
print("Classification rate is :", str((2/3)*100)+"%")
```
